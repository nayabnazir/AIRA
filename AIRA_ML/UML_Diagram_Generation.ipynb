```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
def extract_uml_components(requirement):
    
    words = word_tokenize(requirement.lower())
    tagged = pos_tag(words)
    
    actor = None
    action = []
    obj = []
    
    for word, tag in tagged:
        
        # Actor
        if word in ['user', 'admin']:
            actor = word.capitalize()
        
        # Verb (action)
        elif tag.startswith('VB'):
            action.append(word)
        
        # Noun (object)
        elif tag in ['NN', 'NNS']:
            obj.append(word)
    
    action_text = " ".join(action[:2])  # limit for clean output
    object_text = " ".join(obj[:2])
    
    usecase = f"{action_text} {object_text}".strip()
    
    return actor, usecase


# Test
req = "User can add products to cart"
print(extract_uml_components(req))
```

```python
def generate_smart_uml(requirements_list):
    
    uml = "@startuml\n"
    
    actors_set = set()
    usecases = []
    
    for req in requirements_list:
        actor, usecase = extract_uml_components(req)
        
        if actor:
            actors_set.add(actor)
        
        if usecase:
            usecases.append((actor, usecase))
    
    # Actors
    for actor in actors_set:
        uml += f"actor {actor}\n"
    
    uml += "\nrectangle System {\n"
    
    for _, uc in usecases:
        uml += f"  ({uc})\n"
    
    uml += "}\n\n"
    
    # Relationships
    for actor, uc in usecases:
        if actor:
            uml += f"{actor} --> ({uc})\n"
    
    uml += "@enduml"
    
    return uml


# Test with multiple requirements
requirements = [
    "User can add products to cart",
    "User can place order",
    "Admin can manage products"
]

print(generate_smart_uml(requirements))
```
